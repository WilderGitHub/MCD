```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
aux = pd.read_excel("borrar.xlsx")
aux=aux.drop(columns=["Unnamed: 0"])
aux
```

```python
aux1 = aux.loc[:,"GESTION":"TNT"]
aux2 = aux.loc[:,"KILBRU":"VALOR"]
aux2 = aux2.drop(columns=["KILNET","FINUNI"])
aux2["SUMA"]=aux2["KILBRU"]+aux2["VALOR"]
aux2
```

```python
#esto es para exportar dataframe a excel
#df.to_excel('borrar.xlsx')
```

```python
# para borrar los últimos 10 registros por ejemplo. Para borrar los n primeros seria df.tail(-n)
aux3=aux2.head(-10)
```

```python
# Para concatenar pero al ladito, no debajito. AL LADITO
pd.concat([aux1,aux3],axis=1)
```

```python
#esto es para concatenar de bajito. Nótese como lo pone abajo y a un ladito. Cachas?
#abajo y aun ladito. Because no hay columnas con el mismo nombre pe
pd.concat([aux1,aux3])
```

```python
# otra forma es JOIN, no se bien todavía en que se diferencia con el anterior oe.
auxañadido = DataFrame.join(aux1,aux3)
auxañadido
```

```python
# Este otro es para añadir registros abajo del DataFrame y que reindexe todo. 
# Con el False, mantiene el indexado original
añadidoplus=auxañadido.append(auxañadido, ignore_index=True)
añadidoplus
```
