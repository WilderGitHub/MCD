```python
df = df.groupby(['price', 'query']).size()
df.head(20)
```

```python
temp = df.unstack(level=-1)
temp
```

```python
kaggle = temp.reset_index().fillna(0).set_index('price')
```

```python
kaggle.head()
```

```python
kaggle.columns
```

```python
kaggle.loc[0,"acoustic guitar clamp"]
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
kaggle_sets = kaggle.applymap(encode_units)
kaggle_sets
```

```python
frequent_itemsets = apriori(kaggle_sets, min_support=0.001, use_colnames=True) # min_support = 0.07
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
```
