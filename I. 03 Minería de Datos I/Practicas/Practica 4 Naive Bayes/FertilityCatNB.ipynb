```python
import numpy as np
import pandas as pd

#import urllib.request

import sklearn
from mixed_naive_bayes import MixedNB
#from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split

from sklearn import metrics
from sklearn.metrics import accuracy_score

# Import LabelEncoder
from sklearn import preprocessing
```

```python
dataset = pd.read_csv('Fertility1.csv', sep=';')
```

```python
#labelEncoder
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
dataset['season_cd']=le.fit_transform(dataset.loc[:, 'season'])
dataset['Ch_Disease_cd']=le.fit_transform(dataset.loc[:, 'Ch_Disease'])
dataset['Accident_cd']=le.fit_transform(dataset.loc[:, 'Accident'])
dataset['Surg_Int_cd']=le.fit_transform(dataset.loc[:, 'Surg_Int'])
dataset['H_Fever_cd']=le.fit_transform(dataset.loc[:, 'H_Fever'])
dataset['Alcohol_Cons_cd']=le.fit_transform(dataset.loc[:, 'Alcohol_Cons'])
dataset['Smoke_cd']=le.fit_transform(dataset.loc[:, 'Smoke'])
dataset['Diag_cd']=le.fit_transform(dataset.loc[:, 'Diag'])
```

```python
dataset = dataset.drop(['season', 'Ch_Disease', 'Accident', 'Surg_Int', 'H_Fever', 'Alcohol_Cons', 'Smoke', 'Diag'], axis=1)
```

```python
#X = pd.array(dataset.iloc[:,0:5])
X_train = dataset.iloc[0:80, 0:9].to_numpy()
y_train = dataset.iloc[0:80, 9].to_numpy()

X_test = dataset.iloc[81:, 0:9].to_numpy()
y_test = dataset.iloc[81:, 9].to_numpy()
```

```python
clf = MixedNB(categorical_features=[2, 3, 4, 5, 6, 7, 8])
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred))
```
