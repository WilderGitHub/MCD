```python
import numpy as np
import pandas as pd

#import urllib.request

import sklearn
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
#from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split

from sklearn import metrics
from sklearn.metrics import accuracy_score

# Import LabelEncoder
from sklearn import preprocessing
```

```python
dataset = pd.read_csv('Fertility1.csv', sep=";")
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
#labelEncoder
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
dataset['season_cd']=le.fit_transform(dataset.loc[:, 'season'])
dataset['Ch_Disease_cd']=le.fit_transform(dataset.loc[:, 'Ch_Disease'])
dataset['Accident_cd']=le.fit_transform(dataset.loc[:, 'Accident'])
dataset['Surg_Int_cd']=le.fit_transform(dataset.loc[:, 'Surg_Int'])
dataset['H_Fever_cd']=le.fit_transform(dataset.loc[:, 'H_Fever'])
dataset['Alcohol_Cons_cd']=le.fit_transform(dataset.loc[:, 'Alcohol_Cons'])
dataset['Smoke_cd']=le.fit_transform(dataset.loc[:, 'Smoke'])
```

```python
dataset.head()
```

```python
dataset = dataset.drop(['season', 'Ch_Disease', 'Accident', 'Surg_Int', 'H_Fever', 'Alcohol_Cons', 'Smoke'], axis=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
X = dataset.loc[:, dataset.columns != 'Diag']
y = dataset.loc[:, dataset.columns == 'Diag']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
y
```

```python
BernNB = BernoulliNB(binarize = 0.1)
BernNB.fit(X_train, y_train)
print(BernNB)
```

```python
y_expect = y_test
y_pred = BernNB.predict(X_test)
print(accuracy_score(y_expect, y_pred))
```

```python
MultiNB = MultinomialNB()
MultiNB.fit(X_train, y_train)
print(MultiNB)
y_pred = MultiNB.predict(X_test)
print(accuracy_score(y_expect, y_pred))
```

```python
GaussNB = GaussianNB()
GaussNB.fit(X_train, y_train)
print(GaussNB)
y_pred = GaussNB.predict(X_test)
print(accuracy_score(y_expect, y_pred))
```
